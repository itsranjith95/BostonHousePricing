```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
```

```python
dataset = pd.read_csv("BostonHousing.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
# Summarizing the stats of the data

dataset.describe()
```

```python
# Checking the missing values

dataset.isnull().sum()
```

```python
dataset = dataset.dropna(subset=['rm'])
```

```python
dataset.isnull().sum()
```

```python
# Exploratory Data Analysis

#correlation
dataset.corr()
```

```python
dataset.columns = dataset.columns.str.strip()  # removes spaces
dataset.rename(columns={'medv': 'Price'}, inplace=True)
```

```python
dataset.head()
```

```python
sns.pairplot(dataset)
```

```python
plt.scatter(dataset['crim'], dataset['Price'])
plt.xlabel("CRIME RATE")
plt.ylabel("Price")
```

```python
plt.scatter(dataset['rm'], dataset['Price'])
plt.xlabel("CRIME RATE")
plt.ylabel("Price")
```

```python
sns.regplot(x='rm', y="Price", data=dataset)
```

```python
sns.regplot(x='lstat', y="Price", data=dataset)
```

```python
sns.regplot(x='ptratio', y="Price", data=dataset)
```

```python
## Independent and dependent features

X = dataset.iloc[ : , :-1]
y = dataset.iloc[ : , -1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
print((X_train.shape) ,
     (X_test.shape) ,
     (y_train.shape) ,
     (y_test.shape))
```

```python
## Standardizing the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python

import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

```python
X_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
#Coefficients

print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
# plot the scatter plot for the prediction

plt.scatter(y_test, reg_pred)
```

```python
## Residuals

residual = y_test - reg_pred
```

```python
residual
```

```python
## Plot this residuals

sns.displot(residual, kind='kde')
```

```python
## Scatter plot wrt predtiction and residuals

plt.scatter(reg_pred, residual)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print("mean_squared_error : ", mean_squared_error(y_test, reg_pred))
print("mean_absolute_error : ",mean_absolute_error(y_test, reg_pred))

print("Root square : ", np.sqrt(mean_absolute_error(y_test, reg_pred)))
```

```python
# R sqaure and Adjusted R square

from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
```

```python
score
```

```python
# Adjusted R Square

1 - (1-score) * (len(y_test)-1) / ((len(y_test)-1)-X_test.shape[1]-1)
```
